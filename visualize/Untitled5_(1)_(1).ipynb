```python
embed_moose = np.load("embed_moose.npy")
print("Shape of embed_moose:", embed_moose.shape)
```

```python
embed_time = np.load("embed_timesformer.npy")
print("Shape of embed_time:", embed_time.shape)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(
    n_components=2,            # Chiếu xuống 2D để vẽ
    perplexity = 30,           # Hoặc 40, 50 (nằm trong khoảng 5-50)
    early_exaggeration=12.0,   # Giữ mặc định (hoặc 8.0 ~ 12.0)
    learning_rate='auto',
    max_iter= 4000,# Hoặc khoảng [100..1000], tùy số lượng mẫu             # Tăng số vòng lặp để kết quả ổn định
    n_iter_without_progress=300,
    metric='cosine',
    init='pca',                # Thường cho kết quả ổn định hơn 'random'
    method='barnes_hut',       # Phù hợp cho số mẫu lớn (O(NlogN))
    angle=0.5,                 # 0.2~0.8, nhỏ hơn thì chính xác hơn nhưng chậm
    random_state=42,           # Cố định seed để có thể tái lập
    verbose=1,
    n_jobs = -1
)
```

```python
embeddings_2d_time = tsne.fit_transform(embed_time)
```

```python
embeddings_2d_moose = tsne.fit_transform(embed_moose)
```

```python
embeddings_2d_vivit = tsne.fit_transform(embed_vivit)
```

```python
embeddings_2d_mae = tsne.fit_transform(embed_mae)
```

```python
import matplotlib.pyplot as plt

unique_labels = sorted(list(set(loaded_labels)))
cmap = plt.get_cmap('gist_ncar', len(unique_labels))
color_map = {label: cmap(idx) for idx, label in enumerate(unique_labels)}
```

```python
def viz(embed):
  scale_factor = 2
  embeddings_scaled = embed * scale_factor

  # 2) Tăng kích thước figure
  fig, ax = plt.subplots(figsize=(10, 8))

  ax.axis("off")
  ax.set_facecolor('black')  # Nền đen giúp màu sắc nổi bật

  # 5) Plot từng lớp với kích thước điểm nhỏ, alpha cao, edgecolor='none'
  for idx, lab in enumerate(unique_labels):
      indices = [i for i, l in enumerate(loaded_labels) if l == lab]
      ax.scatter(
          embeddings_scaled[indices, 0],
          embeddings_scaled[indices, 1],
          s=5,                    # Kích thước điểm
          alpha=0.8,              # Độ trong suốt
          color=color_map[lab],
          edgecolors='none'
      )
  plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

scale_factor = 2
embeddings_scaled = embeddings_2d_vivit * scale_factor

# 2) Tăng kích thước figure
fig, ax = plt.subplots(figsize=(10, 8))

ax.axis("off")
ax.set_facecolor('black')  # Nền đen giúp màu sắc nổi bật

# 5) Plot từng lớp với kích thước điểm nhỏ, alpha cao, edgecolor='none'
for idx, lab in enumerate(unique_labels):
    # Fix: Filter indices to be within the bounds of embeddings_scaled
    indices = [i for i, l in enumerate(loaded_labels) if l == lab and i < embeddings_scaled.shape[0]]

    # Skip if indices is empty to avoid error with scatter
    if not indices:
        continue

    ax.scatter(
        embeddings_scaled[indices, 0],
        embeddings_scaled[indices, 1],
        s=10,                    # Kích thước điểm
        alpha=0.8,              # Độ trong suốt
        color=color_map[lab],
        edgecolors='none'
    )
plt.title("t-SNE Visualization", color='white', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scale factor for visualization
scale_factor = 2
embeddings_scaled = embeddings_2d_moose * scale_factor

# Create a figure with a black background
fig, ax = plt.subplots(figsize=(10, 8))
ax.axis("off")
ax.set_facecolor('black')  # Black background for better contrast

# Plot each class with unique colors
for idx, lab in enumerate(unique_labels):
    # Filter indices for the current label
    indices = [i for i, l in enumerate(loaded_labels) if l == lab and i < embeddings_scaled.shape[0]]
    
    # Skip if no indices are found
    if not indices:
        continue
    
    # Scatter plot for the current label
    ax.scatter(
        embeddings_scaled[indices, 0],
        embeddings_scaled[indices, 1],
        s=10,                    # Point size
        alpha=0.8,               # Transparency
        color=color_map[lab],    # Color from the color map
        edgecolors='none'        # No edge color
    )

# Add a title to the plot
plt.title("t-SNE Visualization of Moose Embeddings", color='white', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def viz(embed1, embed2, title1="Embedding 1", title2="Embedding 2"):
  scale_factor = 2
  embed1_scaled = embed1 * scale_factor
  embed2_scaled = embed2 * scale_factor

  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))  # 1 row, 2 columns

  # Embedding 1
  ax1.axis("off")
  ax1.set_facecolor('black')
  for idx, lab in enumerate(unique_labels):
      indices = [i for i, l in enumerate(loaded_labels) if l == lab]
      ax1.scatter(
          embed1_scaled[indices, 0],
          embed1_scaled[indices, 1],
          s=5,
          alpha=0.8,
          color=color_map[lab],
          edgecolors='none'
      )
  ax1.set_title(title1, color='white', fontsize=16)

  # Embedding 2
  ax2.axis("off")
  ax2.set_facecolor('black')
  for idx, lab in enumerate(unique_labels):
      indices = [i for i, l in enumerate(loaded_labels) if l == lab]
      ax2.scatter(
          embed2_scaled[indices, 0],
          embed2_scaled[indices, 1],
          s=5,
          alpha=0.8,
          color=color_map[lab],
          edgecolors='none'
      )
  ax2.set_title(title2, color='white', fontsize=16)

  plt.show()

viz(embeddings_2d_time, embeddings_2d_time, "Moose Embeddings", "Timesformer Embeddings")
```
